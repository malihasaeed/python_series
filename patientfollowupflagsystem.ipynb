```python
#Print a number grid and highlight even numbers with *
for i in range (0,20):
  if i%2==0:
    i="*"
    print(i)
  else:
    print(i)
```

```python
#if patient's ID is an even_number-they need follow up.
def flag_followup(patient_ids):
    for pid in patient_ids:
        if pid % 2 == 0:
            print(pid, "- FOLLOW-UP REQUIRED")
        else:
            print(pid, "- Normal Checkup")

user_input = input("Enter patient IDs separated by commas: ")
ids_list = [int(x) for x in user_input.split(",")]

flag_followup(ids_list)
```
